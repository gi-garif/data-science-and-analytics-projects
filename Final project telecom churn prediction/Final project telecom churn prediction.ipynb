# Project Description

A telecommunication services provider "NoDisconnection.com" wants to learn how to forecast customer churn. If it is discovered that a user plans to leave, they will be offered promo codes and special conditions. The operator's team has collected personal data about certain customers, information about their tariffs and contracts.

**Client's Services Description:**

The operator provides two main types of services:
1. Fixed-line telephone communication: users can connect their telephone devices to multiple lines simultaneously.
2. Internet: connection can be of two types: through a telephone line (DSL - digital subscriber line) or via fiber optic cable (Fiber optic).

Additionally, the following services are available:
- Internet security: antivirus (DeviceProtection) and blocking of unsafe websites (OnlineSecurity).
- Dedicated technical support line (TechSupport).
- Cloud file storage for data backup (OnlineBackup).
- Streaming television (StreamingTV) and a movie catalog (StreamingMovies).

Customers can pay for the services on a monthly basis or sign a contract for 1-2 years. Various payment methods are available, and customers have the option to receive electronic invoices.

**Data Description:**

The data consists of files obtained from various sources:
- contract_new.csv: information about contracts.
- personal_new.csv: personal data of customers.
- internet_new.csv: information about internet services.
- phone_new.csv: information about telephone services.

In all files, the column "customerID" contains the customer's code.

The contract information is up-to-date as of February 1, 2020.

**To achieve the objective of developing a robust model for accurate customer churn prediction, we will follow a systematic approach consisting of the following essential steps:**

1. **Initial Data Preprocessing:** We will start by carefully preparing the data, which involves thorough data cleaning to create relevant predictors and feature engineering: data encoding to handle categorical variables and scaling to ensure consistent measurements.

2. **Exploratory Data Analysis (EDA):** A comprehensive EDA will be conducted to gain meaningful insights from the data. We will visualize key patterns, trends and relationships to better understand the characteristics of the dataset, potentially uncovering valuable information for our predictive model.

3. **Machine Learning Model Research and Selection:** We will explore various machine learning algorithms suitable for customer churn prediction, focusing on evaluating their performance using the ROC-AUC metric. This metric is crucial for assessing the model's ability to distinguish between churn and non-churn instances effectively.

4. **Testing of the Best Model:** After identifying the most promising model through evaluation, we will rigorously test it using an appropriate validation strategy (e.g., cross-validation) to ensure its robustness and generalizability.

5. **Results Analysis and Conclusions:** In this final step, we will thoroughly analyze the model's predictions, examining its strengths and weaknesses. This will help us draw valuable conclusions about its overall effectiveness in predicting customer churn. Furthermore, any insights gained during the process can guide us in making informed business decisions.

By following this well-structured approach, we aim to develop a high-performing model that empowers the organization to make proactive and data-driven decisions, ultimately reducing customer churn and enhancing overall customer satisfaction.

# Preparing data for model training

```python
# splitting dataframe into train and test samples
target = correlated_df['exited']
features = correlated_df.drop(['exited'] , axis=1)

features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.25, random_state=RANDOM_STATE, stratify=target, shuffle=True)

print(features_train.shape)
print(features_test.shape)
```

```python
# applying OHE encoding to categorical features
# scaling numeric features
numeric_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore', drop='first')

# defining types
numeric_cols = ['tenure', 'total_charges', 'monthly_charges']
categorical_cols = ['type', 'paperless_billing', 'payment_method', 'senior_citizen', 'partner', 'online_backup', 'device_protection', 'streaming_tv', 'streaming_movies', 'multiple_lines']

# initializing ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_cols),
        ('cat', categorical_transformer, categorical_cols)
    ]
)

# saving the results
features_train_transformed = pd.DataFrame(preprocessor.fit_transform(features_train))
features_test_transformed = pd.DataFrame(preprocessor.transform(features_test))
```

```python
features_train_transformed.head()
```

```python
features_test_transformed.head()
```

```python
# making sure that no missing values remain
features_train_transformed.isna().mean().round(4) * 100
```

```python
features_test_transformed.isna().mean().round(4) * 100
```

```python
# replacing missing values with median values
features_train_transformed[1].fillna(features_train_transformed[1].median(), inplace=True)
features_test_transformed[1].fillna(features_test_transformed[1].median(), inplace=True)
```

# Model selection

```python
# setting scoring
scoring = {"AUC": "roc_auc", "Accuracy": make_scorer(accuracy_score)}
```

```python
# creating results dataframe
results = {"Model": ["Logistic Regression", "Random Forest", "KNN", "CatBoost", "Gaussian Naive Bayes", "SVM"], "ROC-AUC": [], "Training time": []}
```

## Logistic Regression

```python
param_grid = {}

# initializing the model
model_lr = LogisticRegression(random_state=RANDOM_STATE, class_weight='balanced')

# setting params for RandomizedSearchCV
lr_random_search = RandomizedSearchCV(estimator=model_lr, param_distributions=param_grid,
                                   scoring=scoring, refit="AUC", return_train_score=True, cv=5,
                                      random_state=RANDOM_STATE)

start_time = time.time()

lr_random_search.fit(features_train_transformed, target_train)

end_time = time.time()
execution_time = end_time - start_time

best_params_lr = lr_random_search.best_params_
best_model_lr = lr_random_search.best_estimator_

roc_auc = lr_random_search.best_score_

print(f"ROC-AUC Score: {roc_auc:.4f}")
```

```python
# saving the results
results['ROC-AUC'].append(roc_auc)
results['Training time'].append(execution_time)
```

## Random Forest


```python
# setting hyperparams
param_dist = {
    'n_estimators': [80,100,120,150],
    'max_depth': [10,20,30],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2, 4],
    'class_weight': ['balanced']
    }

rf_model = RandomForestClassifier(random_state=RANDOM_STATE)

rf_random_search = RandomizedSearchCV(estimator=rf_model, param_distributions=param_dist,
                                   scoring=scoring, refit="AUC", return_train_score=True,
                                   n_iter=20, cv=5, random_state=RANDOM_STATE)

start_time = time.time()

# training the model
rf_random_search.fit(features_train_transformed, target_train)

end_time = time.time()
execution_time = end_time - start_time

# saving the best estimator
best_rf_model = rf_random_search.best_estimator_

print("Best ROC-AUC Score:", rf_random_search.best_score_)
print("Best params:", rf_random_search.best_params_)
```

```python
# saving the results
results['ROC-AUC'].append(rf_random_search.best_score_)
results['Training time'].append(execution_time)
```

## K-Nearest Neighbors

```python
# setting hyperparams
param_dist = {
    'n_neighbors': range(1, 21),
    'weights': ['uniform', 'distance'],
    'p': [1, 2],
}

knn_classifier = KNeighborsClassifier()

# RandomizedSearchCV
knn_random_search = RandomizedSearchCV(estimator=knn_classifier, param_distributions=param_dist,
                                   scoring='roc_auc', n_iter=20, cv=5, random_state=RANDOM_STATE)

start_time = time.time()

# training
knn_random_search.fit(features_train_transformed, target_train)

end_time = time.time()
execution_time = end_time - start_time

# saving the best estimator
best_knn_model = knn_random_search.best_estimator_

print("Best ROC-AUC Score:", knn_random_search.best_score_)
print("Best params:", knn_random_search.best_params_)
```

```python
# saving the results
results['ROC-AUC'].append(knn_random_search.best_score_)
results['Training time'].append(execution_time)
```

## CatBoost

```python
# converting target values to string type
target_train_str = target_train.astype(str)
```

```python
start_time = time.time()

catboost_model = CatBoostClassifier(random_state=RANDOM_STATE, verbose=0)

# setting hyperparams
param_dist = {
    'iterations': [100, 200, 300],
    'depth': [4, 6, 8],
    'learning_rate': [0.01, 0.05, 0.1],
    'l2_leaf_reg': [1, 3, 5, 7],
    'cat_features': [categorical_cols]
}

# RandomizedSearchCV
cb_random_search = RandomizedSearchCV(estimator=catboost_model, param_distributions=param_dist,
                                   scoring=scoring, refit="AUC", return_train_score=True,
                                   n_iter=20, cv=5, random_state=RANDOM_STATE)

start_time = time.time()

# training
cb_random_search.fit(features_train, target_train_str)

end_time = time.time()
execution_time = end_time - start_time

# saving the best estimator
best_catboost_model = cb_random_search.best_estimator_

print("Best ROC-AUC Score:", cb_random_search.best_score_)
print("Best params:", cb_random_search.best_params_)
```

```python
# saving the results
results['ROC-AUC'].append(cb_random_search.best_score_)
results['Training time'].append(execution_time)
```

## Gaussian Naive Bayes

```python
# saving the results
results['ROC-AUC'].append(nb_random_search.best_score_)
results['Training time'].append(execution_time)
```

**Challenges Encountered During the Project:**
   - **Class Imbalance:** Dealing with class imbalance, where the churned customers were roughly 6 times less than the active ones, posed a challenge. We addressed this issue by using stratify=target and shuffling data during train-test splitting, as well as applying class weights in the models.
   - **Feature Selection:** Selecting appropriate features and transforming them also proved challenging. We employed the phik library to analyze correlations and eliminated non-informative characteristics.

**Features Used for Model Training:**
   - Features used included information about contracts, payment types, charges, additional services used, contract duration, customer age, and marital status: type, paperless_billing, payment_method, monthly_charges, total_charges, senior_citizen, partner, online_backup, device_protection, streaming_tv, streaming_movies, multiple_lines, tenure.

**Final Model and Results:**
   - The chosen final model was CatBoost with hyperparameters tuned using cross-validation via RandomizedSearchCV, and RANDOM_STATE = 310723.
   - Quality of the final model: ROC-AUC = 0.89, Accuracy = 0.89.
   - Hyperparameters: learning_rate: 0.1, l2_leaf_reg: 3, iterations: 300, depth: 8.
